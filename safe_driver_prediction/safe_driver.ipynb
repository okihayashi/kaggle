### NOTE
As we can see at histogram above, the column "target" has a bool value, one or zero however they are inbalanced. The number of 'one' is quit smaller than the number of 'zero'.

So first of all, count the number of 'one' by `data_Y_tmp.sum()` and pick the same number as 'one' from 'zero' by `data_0_cut = data[data['target']==0].sample(data_Y_tmp.sum())`.
After that just concatenate these two dataframe to get the target balanced training data.

```python
data_Y_tmp = data['target']
#data_X = data.drop('target', axis=1)

data_0_cut = data[data['target']==0].sample(data_Y_tmp.sum())
data_1 = data[data['target']==1]

data_cut = pd.concat([data_0_cut, data_1])
print(data_0_cut.shape)
print(data_1.shape)
print(data_cut.shape)
```

```python
data_X = data_cut.drop('target', axis=1)
data_Y = data_cut['target']

print(data_X.shape)
print(data_Y.shape)
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_Y, test_Y  = train_test_split(data_X, data_Y, test_size=0.2, random_state=0)

print(train_X.shape)
print(test_X.shape)
print(train_Y.shape)
print(test_Y.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

tree = RandomForestClassifier()
tree = tree.fit(train_X, train_Y)

prediction_tree = tree.predict(test_X)

print(accuracy_score(test_Y, prediction_tree))
print(classification_report(test_Y, prediction_tree))
```
