```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import sys, os
sys.path.append(os.pardir)

data = pd.read_csv('./train.csv')
data.head(10)
```

```python
data_Y_tmp = data['target']
data_0_cut = data[data['target']==0].sample(data_Y_tmp.sum())
data_1 = data[data['target']==1]
data_cut = pd.concat([data_0_cut, data_1])

sns.distplot(data_cut[data_cut["target"]==1].ps_ind_02_cat, hist=True, kde=False)
sns.distplot(data_cut[data_cut["target"]==0].ps_ind_02_cat, hist=True, kde=False)
plt.show()
```
